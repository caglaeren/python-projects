```python
raw_data ="  AhMeT  yILMAZ ;  23 ;  1.78 ;  ahmetYILMAZ@GMAIL.Com   "
```

### 1-Genel Temizlik

```python
clean_data = raw_data.strip() #Baştaki ve sondaki boşlukları siler
```

```python
clean_data
```

```python
stringAyir = clean_data.split(";") #karaktere göre ayırır
```

```python
stringAyir
```

### 2-İsim Soyisim Düzeltme

```python
isim = stringAyir[0].strip() #kenar boşlukları sildim
isim_lower = isim.lower() #tüm harfleri küçült
isim_title = isim.title() #ilk harfleri büyüt
```

```python
isim_title
```

```python
isim_slicing = isim_title[:] #0. indeksten başla sonuna kadar tamamını al 
```

```python
isim_slicing
```

### 3- Yaş İşleme

```python
#Yaşı 10 yıl sonrası için float olarak hesapla
stringAyir[1] #1. indeksi yani 23'ü verir. " 23"
yas = int(stringAyir[1].strip()) #"23" oldu boşlukları sildi ve int'a çevirdi
yas_yeni = float(yas + 10)
```

```python
yas_yeni
```

### 4-Boy Analizi

```python
#1.78 -> floata çevir
stringAyir[2] #'  1.78 'verir
boy = float(stringAyir[2].strip()) #boşluklar düzeltildi ve float'a çevrildi sonuc metre
boy_cm = boy * 100 # santimetreye çevirdik
```

```python
boy_cm
```

### 5-Email Temizleme

```python
#Hepsini küçük harf yap ve sondaki boşlukları sil
stringAyir[3] # '  ahmetYILMAZ@GMAIL.Com' bunu verir
email_str = stringAyir[3].lower() #hepsini küçük harf yaptık
```

```python
email_str
```

```python
email_strip_str = email_str.strip() #boşlukları sildik
```

```python
#@'in indeksini bulalım
email_at = email_strip_str.find("@") #@ karakterini bulduk
email_at #indeksini verecek @'in
```

```python
#@’dan önceki kullanıcı adının ilk 3 harfini al
email_atden_onceki = email_strip_str[ : email_at] #0.indeksten başla email_at'in indeksine kadar al
```

```python
email_atden_onceki
```

```python
#şimdi 'ahmetyilmaz'in ilk 3 harfini alacağız
email_isim = email_atden_onceki[:3] #0. indeksten başla 3. indekse kadar al
```

```python
email_isim
```

```python
print("Kullanıcı:",isim_slicing)
print("Yaş (10 yıl sonra):", yas_yeni)
print("Boy (cm):", boy_cm)
print("Email kullanıcı kodu:", email_isim)
```

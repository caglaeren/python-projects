#SENARYO:
#Bir üniversitede yapılan **öğrenci memnuniyet anketi** sonuçları Python ile analiz edilecektir.

Her öğrenci:

- 1–5 arası puan verir
- 0 girerse anketten çıkar
- Rastgele üretilmiş veriler analiz edilir

```python
#1 – Rastgele Anket Verisi
import random

scores = []

# scores listesi 20 elemana ulaşana kadar döngü devam eder
while len(scores) < 20:
    score = random.randint(0,6) #0-5 arası random sayı üretir

    if score == 0:
        break  #anketi bitir

    if score == 6:
        continue #geçersiz veridir bu turu atlar başa döner

    scores.append(score) #1-5 arası puanlar eklenir

print("Toplam skor", scores)
```

```python
#2- Veri Kontrolü
#Kaç geçerli puan toplandığı:
gecerliPuan = len(scores)
print(f"Geçerli puan toplamı: {gecerliPuan} ")

#While kullanıldı çünkü kaç adet geçerli puan geleceğini önceden bilmiyoruz. Veri random üretiliyor bu nedenle koşullu döngü kullanılır.
```

```python
#3- Ortalama Hesaplama

#ortalama puanı hesapla round() kullanarak 2 basamak göster:
#Eğer liste boşsa ilkinde 0 girilmişse hata almak istemiyoruz:
#len(scores) -> adeti verecek
if len(scores) > 0:
    topPuan = sum(scores)
    ortalama = topPuan / len(scores)

    #2 basamak 
    sonuc = round(ortalama, 2)
    print(f"Ortalama hesaplandı. Sonuç:{sonuc} ")
else:
    print("Ortalama hesaplanamadı tekrar dene")
```

```python
#4- Min-Max Analizi

#En düşük puan- En yüksek puan bulunacak
if len(scores) > 0:
    minDeger = min(scores)
    maxDeger = max(scores)

    print(f"En düşük puan: {minDeger}")
    print(f"En yüksek puan: {maxDeger}")

else:
    print("Min max analizi yapılamadı.")
```

```python
#5- Memnuniyet Analizi (For+If)

#for döngüsü kullanarak: 4 ve 5-> "Memnun", 3->"Kararsız", 1 ve 2-> "Memnun Değil" ve her kategori için sayaç tut.
memnunSayisi = 0
kararsizSayisi = 0
memnunOlmayan = 0

for score in scores:
    if score == 4 or score ==5:
        memnunSayisi += 1
    elif score == 3:
        kararsizSayisi += 1
    elif score == 1 or score== 2:
        memnunOlmayan += 1
    else:
        print("Puanlamada hata oldu.")
```

```python
#6-  Erken Uyarı Sistemi (Break)

#For içinde arka arkaya 3 adet 1 puan gelirse: "⚠️ Kritik memnuniyetsizlik tespit edildi" yazıp döngüyü bitir
arkaArkayaBir = 0

for score in scores:
    if score == 1:
        arkaArkayaBir +=1
    else:
        #Eğer puan 1 değilse olmayacağı için sayaç sıfırlanır
        arkaArkayaBir = 0

    #Kontrol 
    if arkaArkayaBir == 3:
        print("⚠️ Kritik memnuniyetsizlik tespit edildi")
        break #kritik durum oluştu ve döngüden çıkıyoruz
    else:
        print("Kritik bir memnuniyetsizlik durumu olmadı.")
```

```python
#7- GENEL ÖZET
print(f"Toplam geçerli cevap:{gecerliPuan}")
print(f"Ortalama puan:{sonuc}")
print(f"Min puan: {minDeger}")
print(f"Max puan:{maxDeger}")
print(" ")
print(f"Memnun: {memnunSayisi}")
print(f"Kararsız: {kararsizSayisi}")
print(f"Memnun değil: {memnunOlmayan}")
```

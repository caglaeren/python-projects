# Proje Tanımı:

## PROJENİN AMACI:
Bu proje kullanıcının günlük harcamalarını girerek toplam harcamayı, ortalama harcamayı, en büyük/ en küçük harcamayı hesaplayan basit bir bütçe asistanı oluşturmayı amaçlar.

## Gerçek Hayatta Ne İşe Yarar ? 
- Aylık harcama takibi yapmak
- Bütçe kontrolü sağlamak
- Verileri alırken oluşabilecek hataları engelleyerek kullanıcıya net bir finansal tablo sunmak.

## Kullanılan Python Konuları: 
- Fonksiyonlar
- input / output
- args (Sayısısz argüman)
- Önemli gömülü fonksiyonlar (sum, len, round, max, min)
- Scope (Local / Global)
- Hataları ele alma (try / except)
- Hesap makinesi mantığı


```python
#2- Harcama Alma Fonksiyonu
#Kullanıcıdan kaç harcama gireceğini al ve girilen her değeri listeye ekle. Hatalı girişlerde programı kırma
def get_expenses():
    harcamalar_listesi = [] #harcamaları ekleyeceğimiz liste

    try:
        count = int(input("Kaç harcama girmek istiyorsunuz?: ")) #sayısı bize string vereceği için int'a çevirdik
        
        for c in range(count):
            while True:
                try:
                    miktar = float(input(f"{c+1}. harcama tutarını girin: "))
                    harcamalar_listesi.append(miktar) #listeye ekler
                    break #Başarılı girişte döngüden çık
                except ValueError:
                    print("Lütfen geçerli bir sayı giriniz. Hata veriyor.")
                    
    except ValueError:
        print("Hata. Harcama adedini doğru girin.")
    return harcamalar_listesi #listeyi döner
        
                    
```

```python
# 3- Toplam ve Ortalama Fonksiyonu
#parametre: harcamalar listesi
#Toplam harcama hesaplanacak, ortalama harcama hesaplanacak, round() ile 2 basamak göster

def calculate_summary(harcamalar_listesi):
    try:
        total = sum(harcamalar_listesi)
        ortalama = total / len(harcamalar_listesi)
        return round(total, 2), round(ortalama, 2)
    except ZeroDivisionError: #sıfıra bölünme hatası alırsak
        #liste boşsa len() 0 döner bu da bölme hatası yaratır
        print("Henüz harcama girilmedi.")
        return 0.0, 0.0
```

```python
# 4- *Args ile Esnek Hesaplama
#Girilen tüm değerleri topla, kaç argüman geldiğini yazdır
def flexible_total(*args):
    if len(args) == 0:
        print("Hiç harcama girilmedi.")
        return 0
        
    total = sum(args) #girilen değerlerin toplamı
    girilenArg = len(args) #kaç argüman girildiği

    print(f"Toplam {girilenArg} adet argüman girildi.")
    print(f"Girilen tüm değerlerin toplamı: {total}")

    return total
    
```

```python
# 5- En büyük / En küçük harcama
def find_extremes(harcamalar_listesi):
    try:
        enBuyuk = max(harcamalar_listesi)
        enKucuk = min(harcamalar_listesi)

        #print(f"En pahalı harcama: {enBuyuk}")
        #print(f"En düşük harcama: {enKucuk}")
        return enBuyuk, enKucuk
        
    except ValueError:
        #liste boşsa min ve max ValueError döndürür
        return None, None
```

```python
# 6- Kapsam (Scope) Gösterimi
#Global bir currency = "TL" değişkeni tanımla
# Fonksiyon içinde bu değişkeni okumayı dene
# Değiştirmenin neden çalışmadığını yorumla

currency = "TL"

def ParaBirimi():
    #Global değişkeni okumak serbesttir
    print(f"Global sistem para birimi: {currency}")

    try:
        #Global değişkeni doğrudan değiştirmeye çalışalım
        # değiştirebilmek için global currency şeklinde fonk içinde kullanmamız gerekir yoksa onu local değişken sanır

        #Bunu açarak yazınca hata veriyor.
        #currency = "usd"
        #print(f"Yeni para birimi: {currency}")

        print("Global değişkeni 'global' anahtar kelimesi olmadan değiştiremezsiniz")
        
    except:
        print("Hata verdi. Global değişkeni bildirmeden değiştirmeye çalıştınız.")

ParaBirimi()
print("Fonksiyon sonrası global para birimi: ", currency)
```

```python
# 7- Ana Akış (Main Flow)
#Fonksiyonları sırayla çağır

#Harcamaları alalım:
harcamalar_listesi = get_expenses()

#Toplam ve Ortalama
total,ortalama = calculate_summary(harcamalar_listesi)

#En büyük ve en küçük
enBuyuk, enKucuk = find_extremes(harcamalar_listesi)

#Sonuçlar
if total is not None:
    print(f"Toplam harcama: {total} {currency}")
    print(f"Ortalama harcama: {ortalama} {currency}")
    print(f"En yüksek harcama: {enBuyuk} {currency}")
    print(f"En düşük harcama: {enKucuk} {currency}")

```
